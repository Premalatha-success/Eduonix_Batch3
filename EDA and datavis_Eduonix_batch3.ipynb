### How to drop a column?

```python
### Dropping the cabin column(Missing values count is more than 50%)
df=df.drop("Cabin",axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

### How to deal with missing values?

```python
#### How to deal with numerical?Mean or Median

age_median=df["Age"].median()
age_median
```

```python
df["Age"]=df["Age"].replace(np.nan,age_median)
```

```python
df.isnull().sum()
```

```python
### How to replace categ->mode

mode_emb=df["Embarked"].mode()[0]
mode_emb
```

```python
df["Embarked"]=df["Embarked"].replace(np.nan,mode_emb)
```

```python
df.isnull().sum()
```

### how to drop insignificant columns???

```python
df=df.drop(["PassengerId","Ticket","Name"],axis=1)
```
